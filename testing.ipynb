```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
theta_min, theta_max = -25, 25  # degrees
theta = np.radians(np.linspace(theta_min, theta_max, 500))  # angles in radians
I0 = 1.0  # Initial intensity
mu_anode = 50  # cm^-1
mu_filter = 0.45  # cm^-1
mu_housing = 0.55  # cm^-1
t_filter = 0.25  # cm
d_housing = 0.15  # cm
anode_angle = np.radians(7)  # radians
t_anode = 0.1  # cm

# Analytical formula
cos_theta = np.cos(theta)
d_anode = t_anode / np.cos(theta + anode_angle)
I_final = (
    I0 * cos_theta * 
    np.exp(-mu_anode * d_anode) * 
    np.exp(-mu_filter * t_filter) * 
    np.exp(-mu_housing * d_housing)
)

# Cartesian Plot
plt.figure(figsize=(8, 5))
plt.plot(np.degrees(theta), I_final, label="Angular Distribution")
plt.xlabel("Angle (degrees)")
plt.ylabel("Normalized Intensity")
plt.title("X-ray Angular Distribution in MDCT Machine")
plt.grid(True)
plt.legend()
plt.show()

# Polar Plot
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.plot(theta, I_final, label="Angular Distribution")
ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)
ax.set_title("Polar Representation of Angular Distribution", va='bottom')
plt.legend()
plt.show()
```
